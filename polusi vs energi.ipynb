```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
emisi_co2 = pd.read_csv('data rapih\data CO2 million ton.csv', delimiter=',', index_col='LOCATION')
emisi_co = pd.read_csv('data rapih\data CO thousand ton.csv', delimiter=',', index_col='LOCATION')
emisi_nox = pd.read_csv('data rapih\data NOx thousand ton.csv', delimiter=',', index_col='LOCATION')
energi = pd.read_csv('data rapih\data energi million toe.csv', delimiter=',', index_col='LOCATION')

emisi_co2_time = pd.read_csv('data rapih\data CO2 million ton.csv', delimiter=',', index_col='TIME')
emisi_co_time = pd.read_csv('data rapih\data CO thousand ton.csv', delimiter=',', index_col='TIME')
emisi_nox_time = pd.read_csv('data rapih\data NOx thousand ton.csv', delimiter=',', index_col='TIME')
energi_time = pd.read_csv('data rapih\data energi million toe.csv', delimiter=',', index_col='TIME')

emisi_co2_loctime = pd.read_csv('data rapih\data CO2 million ton.csv', delimiter=',', index_col=['LOCATION', 'TIME'])
emisi_co_loctime = pd.read_csv('data rapih\data CO thousand ton.csv', delimiter=',', index_col=['LOCATION', 'TIME'])
emisi_nox_loctime = pd.read_csv('data rapih\data NOx thousand ton.csv', delimiter=',', index_col=['LOCATION', 'TIME'])
energi_loctime = pd.read_csv('data rapih\data energi million toe.csv', delimiter=',', index_col=['LOCATION', 'TIME'])

tahun = emisi_co2['TIME']['AUS']
```

## DATA RATA-RATA DI SELURUH DUNIA

```python
rata2_co = pd.DataFrame(emisi_co_time['Value'][:28])
rata2_nox = pd.DataFrame(emisi_co_time['Value'][:28])
for i in range (28):
    rata2_co['Value'][1990+i] = np.average(emisi_co_time['Value'][1990+i])
    rata2_nox['Value'][1990+i] = np.average(emisi_nox_time['Value'][1990+i])
```

```python
plt.figure(figsize=(15,7))
plt.plot(tahun, emisi_co2['Value']['WLD'], label='Penghasilan Emisi CO2(million ton)')
plt.plot(tahun, rata2_co, label='Penghasilan Emisi CO(thousand ton)')
plt.plot(tahun, rata2_nox, label='Penghasilan Emisi NOx(thousand ton)')
plt.plot(tahun, energi['Value']['WLD'], label='Total Bersih Supply Bahan Bakar(Million TOE(Tonne of Oil Equivalent))')

plt.title('Perbandingan Data Rata-Rata Dunia Tahun 1990 - 2017')
plt.xlabel('Tahun')
plt.ylabel('Nilai')

plt.grid(True)
plt.legend()

plt.show()
```

## DATA JUMLAH DI SELURUH DUNIA

```python
total_co2 = pd.DataFrame(emisi_co2_time['Value'][:28])
total_co = pd.DataFrame(emisi_co_time['Value'][:28])
total_nox = pd.DataFrame(emisi_nox_time['Value'][:28])
total_energi = pd.DataFrame(energi_time['Value'][:28])
for i in range (28):
    total_co2['Value'][1990+i] = np.sum(emisi_co2_time['Value'][1990+i])
    total_co['Value'][1990+i] = np.sum(emisi_co_time['Value'][1990+i])
    total_nox['Value'][1990+i] = np.sum(emisi_nox_time['Value'][1990+i])
    total_energi['Value'][1990+i] = np.sum(energi_time['Value'][1990+i])
```

```python
plt.figure(figsize=(15,7))
plt.plot(tahun, total_co2, label='Penghasilan Emisi CO2(million ton)')
plt.plot(tahun, total_co, label='Penghasilan Emisi CO(thousand ton)')
plt.plot(tahun, total_nox, label='Penghasilan Emisi NOx(thousand ton)')
plt.plot(tahun, total_energi, label='Total Bersih Supply Bahan Bakar(Million TOE(Tonne of Oil Equivalent))')

plt.title('Perbandingan Data Total Dunia Tahun 1990 - 2017')
plt.xlabel('Tahun')
plt.ylabel('Nilai')

plt.grid(True)
plt.legend()

plt.show()
```

### Analisis dari data di seluruh dunia

Dari visualisasi data yang dilakukan pada data di seluruh Dunia bisa dilihat bahwa penghasilan gas emisi CO2 berbanding lurus dengan total supply bahan bakar, sedangkan pada gas emisi CO dan NOx berbanding terbalik. Dari sini bisa dianalisis bahwa penggunaan bahan bakar sangat mempengaruhi emisi gas CO2 di udara dimana gas CO2 adalah gas rumah kaca yang dapat menyebabkan terjadinya pemanasan global. Berdasarkan dari berbagai sumber dan dari sumber ruangguru.com serta detik.com disebutkan bahwa pembakaran juga menghasilkan emisi gas CO dan NOx. Gas CO merupakan salah satu gas beracun yang dihasilkan oleh pembakaran tidak sempurna pada suatu mesin, gas CO terbentuk pada saat mesin kekurangan oksigen saat pembakaran terjadi, tetapi dari data didapat emisi gas CO cenderung menurun dan berbanding terbalik dengan supply bahan bakar, dapat diambil analisis bahwa negara-negara di Dunia sudah menyadari tentang bahayanya penghasilan gas CO dan sudah banyak industri otomotif yang menjadikan penggunaan filter knalpot (exhaust manifold) sebagai standar produksi kendaraan di Dunia saat ini, dengan berkembangnya teknologi berbagai macam bahan filter semakin canggih dan ampuh dalam memfilter pembuangan emisi gas. Sebagai contoh yang didapat dari asumsi.co, di Jakarta tahun 2020 akan mewajibkan seluruh kendaraan melakukan uji emisi, di Indonesia juga sudah diterapkan uji emisi kendaraan yang diatur dengan peraturan menteri yang mewajibkan semua industri memproduksi mobil bensin dengan standar emisi Euro 4 yang mengatur ambang batas maksimal emisi sama atau kurang dari CO 1.0 gram/km dan NOx 0.08/km Sedangkan untuk kendaraan bahan bakar diesel CO 0,5 gram/km dan NOx 0,25/km. Kebijakan standar emisi Euro 4 yang diterkan tersebut terlihat sangat memberikan dampak positif pada penghasilan CO dan NOx yang membuatnya cenderung menurun dari tahun ke tahun.

## Analisis dari seluruh negara di Asia Tenggara

Dari seluruh data di wilayah Asia Tenggara bisa dilihat Indonesia merupakan negara penyumbang emisi CO2 terbanyak dengan jumlah penghasilan emisi CO2 sebanyak hampir 500 juta ton dengan kenaikan yang sangat stabil, dengan supply bahan bakar yang selalu meningkat setiap tahunnya. Dapat diambil kesimpulan bahwa di Indonesia penggunaan bahan bakar selalu meningkat setiap tahunnya, dari segala kebijakan transportasi yang ada terutama di wilayah jabodetabek, kurang berdampak terhadap penggunaan bahan bakar, dilihat dari penghasilan emisi dan penggunaan bahan bakar yang setiap tahunnya tetap naik, dilihat dari keadaan nyatanya di Indonesia banyak hukum-hukum yang tidak diterapkan secara tegas. Dan jika dilihat dari penggunaan kereta api di jabodetabek yang dipublish oleh bps.go.id terlihat kereta jabodetabek meningkat pesat seiring berjalannya waktu, tetapi tetap saja untuk menaiki kereta jabodetabek masih harus berdesakan, sehingga mendorong warga untuk tetap menggunakan kendaraan pribadi sehingga emisi yang dihasilkan oleh CO2 serta supply bahan bakar di Indonesia terus meningkat. Sedangkan dari negara-negara Asia Tenggara bisa dilihat pola yang dihasilkan sangat mirip antara supply bahan bakar yang digunakannya dengan emisi CO2 yang mereka hasilkan, terbukti di daerah Asia Tenggara rata-rata masih menggunakan bahan bakar sebagai sumber energi utamanya, karena di setiap negara di Asia Tenggara terlihat semuanya memberikan grafik yang cenderung meningkat.
